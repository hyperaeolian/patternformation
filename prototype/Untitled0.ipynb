```python
def mandelbrot(x, y, max_iters):
    c = complex(x,y)
    z = 0.0j
    for i in range(max_iters):
        z = abs(z)*abs(z) - c
        if (z.real*z.real + z.imag*z.imag) >= 4:
            return i
    return max_iters
```

```python
def render_fractal(xMin, xMax, yMin, yMax, image, iters):
    height = image.shape[0]
    width = image.shape[1]
    
    pixelWidth = (xMax - xMin) / width
    pixelHeight = (yMax - yMin) / height
    
    for x in range(width):
        real = xMin * pixelWidth
        for y in range(height):
            imag = yMin * pixelHeight
            color = mandelbrot(real, imag, iters)
            image[y,x] = color
```

```python
def frac(xMin, xMax, yMin, yMax, image, iters, complexes):
    height = image.shape[0]
    width = image.shape[1]
    pixelWidth = (xMax - xMin) / width
    pixelHeight = (yMax - yMin) / height
    for x in range(width):
        real = complexes[x] + x * pixelWidth
        for y in range(height):
            imag = complexes[y] + y * pixelHeight
            color = mandelbrot(real, imag, iters)
            image[y,x] = color
```

```python
image = np.zeros((1024, 1536), dtype = np.uint8)
render_fractal(-2.0, 1.0, -1.0, 1.0, image, 100) 
imshow(image)
```
